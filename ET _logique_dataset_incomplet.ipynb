```python
from scipy.spatial.distance import hamming
import kmeans1d
```

### DonnÃ©es catÃ©gorielles avec le LabelEncoder

On transforme les donnÃ©es catÃ©gorielles en donnÃ©es numÃ©riques avec un LabelEncoder.

```python
from sklearn.preprocessing import LabelEncoder
label_encoder =LabelEncoder()

df1=df.copy()
df1.drop('Dead', axis=1,inplace=True)
df1['group']= label_encoder.fit_transform(df['group'])
df1['Breed']= label_encoder.fit_transform(df['Breed'])
df1['breed_size']= label_encoder.fit_transform(df['breed_size'])

df1['sex'] = df['sex'].replace(to_replace={'M':1,'F':0})
df1.dtypes
```

On calcule les distances de Hamming et le ET logique entre les features.

```python
for i in range(11):
    for j in range(i,10):
        if i!=j :
            clusters1, centroids1 = kmeans1d.cluster(df1[df1.columns[i]], 2)
            clusters2, centroids2 = kmeans1d.cluster(df1[df1.columns[j]], 2)
            vec_coherence = []
            for k in range(len(clusters1)):
                if clusters1[k] == clusters2[k] == 1:
                    vec_coherence.append(1)
                else:
                    vec_coherence.append(0)
            print("Hamming + ET logique entre "+str(df1.columns[i])+" et "+str(df1.columns[j]) +" : \n", str(hamming(clusters1,clusters2) * len(clusters1)) +"\n", vec_coherence)
```

On calcule les distances de Hamming entre les features et la vÃ©ritÃ© terrain.

```python
df_dead = df['Dead']
dico = {}
clusters2, centroids2 = kmeans1d.cluster(df_dead, 2)
for i in range(10):
    clusters1, centroids1 = kmeans1d.cluster(df1[df1.columns[i]], 2)
    dico[str(df1.columns[i])+" - Dead"] = hamming(clusters1,clusters2) * len(clusters1)
            
{k:v for k, v in sorted(dico.items(), key=lambda item: item[1])}
```
