```python
dico = {}
for i in df:
    dico[i] = 0

for i in df:
    for j in range(len(df[i])):
        if str(df[i][j]) == "nan" and df["Dead"][j] == 0: 
            dico[i] += 1
            
x = dico
x = {k: v for k,v in x.items() if v!=0}
{k: v for k, v in sorted(x.items(), key=lambda item: item[1], reverse=True)}
```

```python
#Filter : Pearson et Spearman
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import seaborn as sb
```

```python
#Affichage de la matrice de corrélation avec la méthode Spearman
#Chaque cellule indique le résultat de la méthode spearmanr(ligne, colonne)

plt.figure(figsize=(20,20))
matrix = df.corr(method="pearson")
sb.heatmap(matrix,annot=True,cmap=plt.cm.Purples)
plt.show()
```

```python
import statistics as st
L = []
for i in range(len(df["Dead"])):
    if df["Dead"][i] == 1:
        L.append(df["Age_dead"][i])
        
print("Moyenne d'âge de la mort des chiots : ", st.mean(L))
print("Médiane d'âge de la mort des chiots : ", st.median(L))
print("Liste des quartiles pour l'âge de la mort des chiots : ", st.quantiles(L))
print("Variance de l'âge de la mort des chiots : ", st.pvariance(L))
print("Ecart-type de l'âge de la mort des chiots : ",st.pstdev(L))
```
